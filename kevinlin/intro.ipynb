Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Read in the sequential dataset from [here](https://docs.google.com/spreadsheets/d/1OfUa546OzV6GA_LOB2LY9rFgs7LeduRQ/edit#gid=1920191272). Fix some typos within the dataset, and convert all the draw data into readable numbers (pandas reads them as strings due to the typos). Preview the dataframe.

```python
sequential = pd.read_csv("results_sequential.csv", skiprows=2)
sequential = sequential.replace(["<1", "<1.00", "<.5", None, np.nan, "1..03", "8.56^", "7.74^"], [0, 0, 0, -1, -1, 1.03, 8.56, 7.74])
cols = ["1st Draw", "2nd Draw", "3rd Draw", "4th Draw", "5th Draw", "6th Draw", "7th Draw", "8th Draw", "9th Draw", "10th Draw", "11th Draw", "3 Minute", "5 Minute", "7 Minute", "10 Minute", "15 Minute", "20 Minute"]
sequential[cols] = sequential[cols].apply(pd.to_numeric, axis=1)
print(sequential.shape)
sequential.head(15)
```

Simple line graph of all the Sequential Draws.

```python
x_axis = [i for i in range(1, 12)]
for i, row in sequential.iterrows():
    plt.plot(x_axis, [row[x] for x in range(2, 13)])
plt.xlabel("Draw #")
plt.ylabel("Lead (ppb)")
plt.title("Lead Measurement over 11 Draws")
plt.show()
```

Simple line graph of all the Timed Draws.

```python
x_axis = [3, 5, 7, 10, 15, 20]
fig, ax = plt.subplots()
for i, row in sequential.iterrows():
    plt.plot(x_axis, [row[x] for x in range(13, 19)])
for i in x_axis:
    plt.text(i, 70, str(i) + "min")
    plt.axvline(x=i, ls="dotted")
plt.xlabel("Time (min)")
plt.ylabel("Lead (ppb)")
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
plt.title("Lead Measurement over Timed Draws")
plt.show()
```
