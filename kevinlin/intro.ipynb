```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sequential = pd.read_csv("results_sequential.csv", skiprows=2)
sequential = sequential.replace(["<1", "<1.00", "<.5", None, np.nan, "1..03", "8.56^", "7.74^"], [0, 0, 0, -1, -1, 1.03, 8.56, 7.74])
cols = ["1st Draw", "2nd Draw", "3rd Draw", "4th Draw", "5th Draw", "6th Draw", "7th Draw", "8th Draw", "9th Draw", "10th Draw", "11th Draw", "3 Minute", "5 Minute", "7 Minute", "10 Minute", "15 Minute", "20 Minute"]
sequential[cols] = sequential[cols].apply(pd.to_numeric, axis=1)
print(sequential.shape)
sequential.head(15)
```

```python
x_axis = [i for i in range(1, 12)]
for i, row in sequential.iterrows():
    plt.plot(x_axis, [row[x] for x in range(2, 13)])
plt.show()
```
