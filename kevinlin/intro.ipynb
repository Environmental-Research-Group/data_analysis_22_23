The dataset contains street names. Below contains code using a geolocator api service to find more details about each street / dataline so we can make plots geographically. The api service rate limits against bots so this chunk of code will take on average 30 minutes to run, however all the data has already been ported to `nominatim_data.txt`.

```python
# geolocator code that extracts location data based off of the scuffed addresses we get from the sequential csv

from geopy import Nominatim
from geopy.extra.rate_limiter import RateLimiter

locator = Nominatim(user_agent="myGeocoder")
#print(locator.geocode("S Dr Martin Luther King Dr, Chicago, Illinois", addressdetails=True).raw.get("address").get("suburb"))
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
sequential["Street"] = [" ".join(address.split()[1:]) + ", Chicago, Illinois, USA" for address in sequential["Address"]]
geodata = []
i = 1

for i in range(1808):
    street = sequential["Street"][i]
    try:
        data = geocode(street, addressdetails=True).raw
    except AttributeError:
        geodata.append("")
        print(str(i) + "/1808 SKIPPED")
        continue
    geodata.append(data)
    print(str(i) + "/1808")
```

We can now combine the data in `nominatim_data.txt` to our dataframe.

```python
nominatim_data = open("nominatim_data.txt", "r")
lines = nominatim_data.readlines()

geodata = []
for line in lines:
    if line == "\n":
        geodata.append(np.nan)
    else:
        geodata.append(line)


sequential["geodata"] = geodata
sequential.head(15)
```

```python
sequential.to_csv("sequential_cleaned+geo.csv", index=False)
```

The Nominatim geo api returns data in JSON format, we can read the json and pull out specific values we want in the code below. However, the json uses single quotes ' not double quotes " so we have to convert everything into a readable format.

```python
import json
import re

sequential = pd.read_csv("sequential_cleaned+geo.csv")
sequential.head(10)

test_str = sequential["geodata"][0].replace("'", '"')
print(test_str)
test_json = json.loads(test_str)
print(test_json.keys())

sequential["geodata"] = [re.sub(r"(?<![a-zA-Z])'|'(?![a-zA-Z])", r'"', str(x)) for x in sequential["geodata"]]
sequential.head(10)
```

```python
place_id = []
osm_type = []
osm_id = []
boundingbox = []
lat = []
lon = []
display_name = []
clazz = []
typ3 = []
importance = []
# under address
road = []
neighbourhood = []
suburb = []

for data in sequential["geodata"]:
    if data == "nan":
        place_id.append(np.nan)
        osm_type.append(np.nan)
        osm_id.append(np.nan)
        boundingbox.append(np.nan)
        lat.append(np.nan)
        lon.append(np.nan)
        display_name.append(np.nan)
        clazz.append(np.nan)
        typ3.append(np.nan)
        importance.append(np.nan)
        road.append(np.nan)
        neighbourhood.append(np.nan)
        suburb.append(np.nan)
    else:
        temp_json = json.loads(data)
        place_id.append(temp_json.get("place_id"))
        osm_type.append(temp_json.get("osm_type"))
        osm_id.append(temp_json.get("osm_id"))
        boundingbox.append(temp_json.get("boundingbox"))
        lat.append(temp_json.get("lat"))
        lon.append(temp_json.get("lon"))
        display_name.append(temp_json.get("display_name"))
        clazz.append(temp_json.get("class"))
        typ3.append(temp_json.get("type"))
        importance.append(temp_json.get("importance"))
        road.append(temp_json.get("address").get("road"))
        neighbourhood.append(temp_json.get("address").get("neighbourhood"))
        suburb.append(temp_json.get("address").get("suburb"))

sequential["place_id"] = place_id
sequential["osm_type"] = osm_type
sequential["osm_id"] = osm_id
sequential["boundingbox"] = boundingbox
sequential["lat"] = lat
sequential["lon"] = lon
sequential["display_name"] = display_name
sequential["class"] = clazz
sequential["type"] = typ3
sequential["importance"] = importance
sequential["road"] = road
sequential["neighbourhood"] = neighbourhood
sequential["suburb"] = suburb

sequential.head(10)
```

```python
sequential.to_csv("sequential_cleaned+geo_cleaned.csv", index=False)
```

```python
crime = pd.read_csv("crime2016.csv")
print(crime.shape)
crime = crime.sort_values(by="Community Area", ignore_index=True)
crime
```

```python
poverty = pd.read_csv("poverty.csv")
print(poverty.shape)
poverty
```

```python
crime_poverty = poverty.copy()
crime_poverty = crime_poverty[:-1]
crime_poverty = crime_poverty.rename(columns={"COMMUNITY AREA NAME":"suburb"})
crime_poverty["CRIMES"] = [int(x.replace(",", "")) for x in crime["Count of Rows (Reported Crime/Reported Crimes)"]]
crime_poverty
```

```python
plt.scatter(crime_poverty["PERCENT HOUSEHOLDS BELOW POVERTY"], crime_poverty["CRIMES"])
m, b = np.polyfit(crime_poverty["PERCENT HOUSEHOLDS BELOW POVERTY"], crime_poverty["CRIMES"], 1)
plt.plot(crime_poverty["PERCENT HOUSEHOLDS BELOW POVERTY"], crime_poverty["PERCENT HOUSEHOLDS BELOW POVERTY"] * m + b)
plt.show()
```

```python
sequential = pd.read_csv("sequential_cleaned+geo_cleaned.csv")
sequential_suburb = sequential.groupby("suburb").mean().reset_index()
sequential_suburb = sequential_suburb.drop(columns=["place_id", "osm_id", "lat", "lon"])
sequential_suburb["avg draws"] = sequential_suburb.iloc[:, 1:12].sum(axis=1) / 11
sequential_suburb["avg minute"] = sequential_suburb.iloc[:, 12:18].sum(axis=1) / 5
sequential_suburb
```

```python
FINAL_DF = pd.merge(crime_poverty, sequential_suburb, on="suburb")
FINAL_DF.to_csv("lead_poverty_crime_avg_neighborhood.csv", index=False)
FINAL_DF
```

```python
df = pd.read_csv("lead_poverty_crime_avg_neighborhood.csv")
df["crime_rate"] = df["CRIMES"] / df["pop"]
df["max_draw"] = np.max(df.iloc[:, 11:23], axis=1)
df
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.scatter3D(df["avg draws"], df["PERCENT HOUSEHOLDS BELOW POVERTY"], df["crime_rate"], color="k")

x_pred = np.linspace(0, 20, 20)
y_pred = np.linspace(0, 50, 20)
xx_pred, yy_pred = np.meshgrid(x_pred, y_pred)
model_viz = np.array([xx_pred.flatten(), yy_pred.flatten()]).T

from sklearn import linear_model
ols = linear_model.LinearRegression()
X = df[["avg draws", "PERCENT HOUSEHOLDS BELOW POVERTY"]].values.reshape(-1, 2)
Y = df["crime_rate"]
model = ols.fit(X, Y)
predicted = model.predict(model_viz)
r2 = model.score(X, Y)

ax.view_init(elev=10, azim=150)
ax.scatter(xx_pred.flatten(), yy_pred.flatten(), predicted, facecolor=(0,0,0,0), s=20, edgecolor='#70b3f0')

ax.set_xlabel("Avg Lead (ppb)")
ax.set_ylabel("% Poverty")
ax.set_zlabel("Crime Rate")
fig.savefig("lead_poverty_crime.png", bbox_inches="tight")
print(r2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
def regplot(x, y, x_label, y_label, title, save):
    model = LinearRegression()
    X = x.values.reshape(-1, 1)
    model.fit(X, y)
    r2 = r2_score(y, model.predict(X))
    plt.scatter(x, y)
    plt.plot(x, model.predict(X), color="red")
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    print(r2)
    plt.savefig(save, bbox_inches="tight")
    plt.show()
```

```python
regplot(df["avg draws"], df["crime_rate"], "Avg Lead (ppb)", "Crime Rate", "Avg Lead vs. Crime Rate", "leadd_crime.png")
```
