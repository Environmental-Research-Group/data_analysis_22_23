```python
df["Mean"] = df[["1st Draw", "2nd Draw", "3rd Draw", "4th Draw", "5th Draw", "6th Draw", "7th Draw", "8th Draw", "9th Draw", "10th Draw", "11th Draw"]].mean(axis = 1)
```

```python
df["suburb"].unique()
```

```python
lst_sub = []
lst_med = []
for suburb in df["suburb"].unique():
    mean_series = df.loc[df["suburb"] == suburb, "Mean"]
    med = mean_series.median()
    lst_sub.append(suburb)
    lst_med.append(med)
```

```python
lst_sub.remove(np.nan)
lst_med.remove(np.nan)
```

```python
lst_sub
```

```python
lead_data = pd.Series(lst_med, index = lst_sub)
lead_data
```

```python
plt.rcParams['figure.figsize'] = (12,9)
plt.xticks(rotation = 90, fontsize = 6.5)
plt.bar(lst_sub, lst_med, width = 0.4)
```
